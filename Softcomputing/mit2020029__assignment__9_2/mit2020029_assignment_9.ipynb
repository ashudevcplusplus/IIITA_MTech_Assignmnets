```python
from matplotlib import pyplot as imported___lib__plotting
from matplotlib.image import imread
import numpy as imported___lib__numpy
import os
imported__dataset__path = 'dataset/'
imported__dataset__dir  = os.listdir(imported__dataset__path)
var__image__width  = 92
var__image__height = 112
```

```python
print('Train Images:')
structure__used__training__tensor   = imported___lib__numpy.ndarray(shape=(70, var__image__height*var__image__width), dtype=imported___lib__numpy.float64)

for i in range(70):
    var__image__one_by_one = imported___lib__plotting.imread(imported__dataset__path + 'training/'+str(i+1)+'.pgm')
    structure__used__training__tensor[i,:] = imported___lib__numpy.array(var__image__one_by_one, dtype='float64').flatten()
    imported___lib__plotting.subplot(10,7,1+i)
    imported___lib__plotting.imshow(var__image__one_by_one, cmap='gray')
    imported___lib__plotting.tick_params(labelleft=False, labelbottom=False, bottom=False,top=False,right=False,left=False, which='both')
    imported___lib__plotting.subplots_adjust(right=1.2, top=2.5)
imported___lib__plotting.show()




print('Test Images:')
structure__used__testing_____tensor   = imported___lib__numpy.ndarray(shape=(30, var__image__height*var__image__width), dtype=imported___lib__numpy.float64)

for i in range(30):
    var__image__one_by_one = imread(imported__dataset__path + 'test/'+str(i+1)+'.pgm')
    structure__used__testing_____tensor[i,:] = imported___lib__numpy.array(var__image__one_by_one, dtype='float64').flatten()
    imported___lib__plotting.subplot(10,3,1+i)
    imported___lib__plotting.imshow(var__image__one_by_one, cmap='gray')
    imported___lib__plotting.subplots_adjust(right=1.2, top=2.5)
    imported___lib__plotting.tick_params(labelleft=False, labelbottom=False, bottom=False,top=False,right=False,left=False, which='both')
imported___lib__plotting.show()
```

# Calculated Mean, mean zero, and co-variance matrix

```python
calculated____mean__of__face = imported___lib__numpy.zeros((1,var__image__height*var__image__width))

for i in structure__used__training__tensor:
    calculated____mean__of__face = imported___lib__numpy.add(calculated____mean__of__face,i)

calculated____mean__of__face = imported___lib__numpy.divide(calculated____mean__of__face,float(len(structure__used__training__tensor))).flatten()






normalised_training_tensor = imported___lib__numpy.ndarray(shape=(len(structure__used__training__tensor), var__image__height*var__image__width))

for i in range(len(structure__used__training__tensor)):
    normalised_training_tensor[i] = imported___lib__numpy.subtract(structure__used__training__tensor[i],calculated____mean__of__face)



    
    
    
    

co__variance___matrix = imported___lib__numpy.cov(normalised_training_tensor)
co__variance___matrix = imported___lib__numpy.divide(co__variance___matrix,60.0)
print('Covariance matrix of X:')
print(co__variance___matrix)

```

# Calculated Eigenvalues , Eigen Vectors, and also Feature Vectors

```python
calculated___eigen___values, calculated____eigen______vectors, = imported___lib__numpy.linalg.eig(co__variance___matrix)

print('Eigenvalues of Cov(X):',calculated___eigen___values)

print('Eigenvectors of Cov(X):')

print(calculated____eigen______vectors)





calculated___eigen____pairs = [(calculated___eigen___values[index], calculated____eigen______vectors[:,index]) for index in range(len(calculated___eigen___values))]

calculated___eigen____pairs.sort(reverse=True)

performed__eigen___values_sorting___for__features  = [calculated___eigen____pairs[index][0] for index in range(len(calculated___eigen___values))]

eigvectors_sort = [calculated___eigen____pairs[index][1] for index in range(len(calculated___eigen___values))]

variance__explained__vector_comp_sum = imported___lib__numpy.cumsum(performed__eigen___values_sorting___for__features)/sum(performed__eigen___values_sorting___for__features)

print("Cumulative proportion of variance explained vector:", variance__explained__vector_comp_sum)





num_comp = range(1,len(performed__eigen___values_sorting___for__features)+1)

imported___lib__plotting.title('Cum. Prop. Variance and Components Kept')

imported___lib__plotting.xlabel('Principal Components')

imported___lib__plotting.ylabel('Cum. Prop. Variance ')

imported___lib__plotting.scatter(num_comp, variance__explained__vector_comp_sum)

imported___lib__plotting.show()




print('Number of eigen vectors:',len(performed__eigen___values_sorting___for__features))

total__number__selected_components = 30

print("k:",total__number__selected_components)

reduced_data = imported___lib__numpy.array(eigvectors_sort[:total__number__selected_components]).transpose()
```

# Eigen Faces and signatures

```python
proj_data = imported___lib__numpy.dot(structure__used__training__tensor.transpose(),reduced_data)
proj_data = proj_data.transpose()
proj_data.shape




print(normalised_training_tensor.shape)
print(proj_data.shape)





w = imported___lib__numpy.array([imported___lib__numpy.dot(proj_data,var__image__one_by_one) for var__image__one_by_one in normalised_training_tensor])
w.shape
```

# Testing

```python

total___image___count=0
total___number__images=0
total__number__of__correct__predictions__count=0


def recogniser(img_number,proj_data,w):
    global total___image___count,highest_min,total___number__images,total__number__of__correct__predictions__count
    
    total___number__images          += 1
    unknown_face_vector = structure__used__testing_____tensor[img_number,:]
    normalised_uface_vector = imported___lib__numpy.subtract(unknown_face_vector,calculated____mean__of__face)
    
    imported___lib__plotting.subplot(11,8,1+total___image___count)
    imported___lib__plotting.imshow(unknown_face_vector.reshape(var__image__height,var__image__width), cmap='gray')
    imported___lib__plotting.title('Input:'+str(img_number+1))
    imported___lib__plotting.subplots_adjust(right=1.2, top=2.5)
    imported___lib__plotting.tick_params(labelleft=False, labelbottom=False, bottom=False,top=False,right=False,left=False, which='both')
    total___image___count+=1
    
    w_unknown = imported___lib__numpy.dot(proj_data, normalised_uface_vector) 
    diff  = w - w_unknown
    array___norms___values = imported___lib__numpy.linalg.norm(diff, axis=1)
    index = imported___lib__numpy.argmin(array___norms___values)
    
    imported___lib__plotting.subplot(11,8,1+total___image___count)
    
    variable__set___number = int(img_number/4)
    t0 = 15000000
       
    if array___norms___values[index] < t0:
        if(index>=(6*variable__set___number) and index<(6*(variable__set___number+1))):
            imported___lib__plotting.title('Matched with:'+str(index+1), color='g')
            imported___lib__plotting.imshow(structure__used__training__tensor[index,:].reshape(var__image__height,var__image__width), cmap='gray')
            total__number__of__correct__predictions__count += 1
        else:
            imported___lib__plotting.title('Matched with:'+str(index+1), color='r')
            imported___lib__plotting.imshow(structure__used__training__tensor[index,:].reshape(var__image__height,var__image__width), cmap='gray')
    else:
        if(img_number>=40):
            imported___lib__plotting.title('Un-detected face!', color='b')
            total__number__of__correct__predictions__count += 1
        else:
            imported___lib__plotting.title('Un-detected face!', color='y')
    imported___lib__plotting.subplots_adjust(right=1.2, top=2.5)
    imported___lib__plotting.tick_params(labelleft=False, labelbottom=False, bottom=False,top=False,right=False,left=False, which='both')
    total___image___count+=1

fig = imported___lib__plotting.figure(figsize=(10, 10))
for i in range(len(structure__used__testing_____tensor)):
    recogniser(i,proj_data,w)

imported___lib__plotting.show()

print('Correct predictions: {}/{} = {}%'.format(total__number__of__correct__predictions__count, 30, total__number__of__correct__predictions__count/30*100.00))

```
