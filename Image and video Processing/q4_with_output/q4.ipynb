```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow


imported_image = cv2.imread("/content/Figure8.png")

gray_______converted_________image = cv2.cvtColor(imported_image, cv2.COLOR_BGR2GRAY)
imported_image_____blurred = cv2.GaussianBlur(gray_______converted_________image, (17, 17), 0)

cv2_imshow( gray_______converted_________image)
cv2_imshow( imported_image_____blurred)
cv2.waitKey(0)
imported_image________outline = cv2.Canny(imported_image_____blurred, 30, 150)

cv2_imshow(imported_image________outline)
cv2.waitKey(0)

(cnts, _) = cv2.findContours(imported_image________outline, cv2.RETR_EXTERNAL,
            cv2.CHAIN_APPROX_SIMPLE)

cv2.drawContours(imported_image, cnts, -1, (0, 255, 0), 2)
cv2_imshow(imported_image)
cv2.waitKey(0)
print("Number of coins found %i " % len(cnts))
```
