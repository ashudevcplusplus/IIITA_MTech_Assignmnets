# Find the mean value of every attribute and replace with NaN in df1 ?

```python
df_1 = df.iloc[:split_50_51,:] 
df_2 = df.iloc[split_50_51 + 1:,:] 
```

```python
mean_value=df_1['Production'].mean()
```

```python
print(mean_value)
```

```python
df_1=df_1.fillna(mean_value)

details = pd.DataFrame(df_1, columns =['State_Name','District_Name','Crop_Year','Season','Crop','Area','Production'], ) 
             
print(" \nshow the boolean Dataframe :\n", details.isna()) 
  
print("Count total NaN at each column in a DataFrame:\n",details.isnull().sum()) 
```
